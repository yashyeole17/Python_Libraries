```python
import matplotlib.pyplot as plt
x=[2,5,8,10,12,15,20,25,37,40]
y=[10,20,30,40,50,67,76,89,91,100]
plt.figure(figsize=(5,5))
plt.xlabel('values for "X" axis')
plt.ylabel('Budget amount')
plt.title('GRAPH for AVERAGE v/s Month')
plt.grid()
plt.xlim(0,50)
plt.ylim(0,100)
plt.plot(x,y,color='g',marker='.',label='2019')
plt.legend()
```

```python
 plt.figure(figsize=(5,5))
plt.xlabel('values for "X" axis')
plt.ylabel('Budget amount')
plt.title('GRAPH for AVERAGE v/s Month')
plt.grid()
plt.xlim(0,40)
plt.ylim(0,100)
x=[2,5,8,10,12,15,20,25,37,40]
y=[10,20,30,40,50,67,76,89,91,100]
z=[15,25,45,45,50,69,77,80,85,90]
t=[10,23,47,50,60,70,75,84,92,99]
plt.plot(x,y,color='g',marker='.',label='2019')
plt.plot(x,z,color='r',marker='*',label='2020',linestyle='dashed')
plt.plot(x,t,color='b',marker='+',label='2021',linestyle='-.')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
index = np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
bw = 0.3
plt.axis([0,5,0,8])
plt.title('A Multiseries Bar Chart',fontsize=20)
plt.bar(index,values1,bw,color='b')
plt.bar(index+bw,values2,bw,color='g')
plt.bar(index+2*bw,values3,bw,color='r')
#plt.xticks(index+1.5*bw,['A','B','C','D','E'])
```
